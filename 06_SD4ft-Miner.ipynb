# Pokud má každý hráč CT helmu, jaké vybavení ovlivňuje jejich výhru, když T strana vlastní nebo nevlastní více jak 3 zbraně AK-47?

```python
#imports of required packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from cleverminer import cleverminer
```

```python
#loading of dataset
df = pd.read_csv('./data/csgo_round_snapshots_processed.csv', sep=',')
df_one = df.copy(deep=True)
df_two = df.copy(deep=True)
```

```python
pd.set_option('display.max_columns', None)
```

```python
#exploring data
df.head()
```

```python
df_two = df_two[df_two['ct_helmets'] == 5]
```

```python
df_two['t_weapon_ak47'] = df_two['t_weapon_ak47'] > 3 
```

```python
df_two = df_two.loc[:, ['round_winner', 't_weapon_ak47', 'ct_category_sniper_amount',
                   'ct_category_rifle_amount', 'ct_category_shotgun_amount','ct_category_grenade_amount',
                        'ct_category_armor_price', 'ct_category_smg_amount']]
```

```python
clm_two = cleverminer(df=df_two,proc='SD4ftMiner',
               quantifiers= {'Base1':50, 'Base2':50, 'Ratioconf' : 1.7},
               ante ={
                    'attributes':[
                        {'name': 'ct_category_sniper_amount', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'ct_category_rifle_amount', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'ct_category_shotgun_amount', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'ct_category_smg_amount', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'ct_category_armor_price', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'ct_category_grenade_amount', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':2, 'type':'con'},
               succ ={
                    'attributes':[
                        {'name': 'round_winner', 'type': 'one', 'value': 'CT'}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
               frst ={
                    'attributes':[
                        {'name': 't_weapon_ak47', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
               scnd ={
                    'attributes':[
                        {'name': 't_weapon_ak47', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'}
               )


clm_two.print_summary()
clm_two.print_rulelist()
clm_two.print_rule(3)
```
