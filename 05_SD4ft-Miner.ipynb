```python
#imports of required packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from cleverminer import cleverminer
```

```python
#loading of dataset
df = pd.read_csv('./data/csgo_round_snapshots_processed.csv', sep=',')
df_one = df.copy(deep=True)
df_two = df.copy(deep=True)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_one = df_one[(df_one['ct_players_alive'] == 1) & (df_one['t_players_alive'] == 1)]
```

```python
df_one = df_one.loc[:, ['round_winner', 'ct_defuse_kits', 'ct_category_pistol_amount', 'ct_category_sniper_amount',
                   'ct_category_rifle_amount', 'ct_category_shotgun_amount','ct_category_grenade_amount', 'ct_category_smg_amount',
                   't_category_pistol_amount', 't_category_sniper_amount', 't_category_smg_amount',
                   't_category_rifle_amount', 't_category_shotgun_amount','t_category_grenade_amount']]
```

```python

clm_one = cleverminer(df=df_one,proc='SD4ftMiner',
               quantifiers= {'Base1':25, 'Base2':25, 'Ratioconf' : 1.5},
               ante ={
                    'attributes':[
                        {'name': 'ct_category_sniper_amount', 'type': 'one', 'value': '1.0'},
                        {'name': 'ct_category_rifle_amount', 'type': 'one', 'value': '1.0'},
                        {'name': 'ct_category_shotgun_amount', 'type': 'one', 'value': '1.0'},
                        {'name': 'ct_category_smg_amount', 'type': 'one', 'value': '1.0'},
                        {'name': 'ct_category_grenade_amount', 'type': 'one', 'value': '1.0'},
                        {'name': 't_category_sniper_amount', 'type': 'one', 'value': '1.0'},
                        {'name': 't_category_rifle_amount', 'type': 'one', 'value': '1.0'},
                        {'name': 't_category_smg_amount', 'type': 'one', 'value': '1.0'},
                        {'name': 't_category_grenade_amount', 'type': 'one', 'value': '1.0'}
                    ], 'minlen':1, 'maxlen':3, 'type':'con'},
               succ ={
                    'attributes':[
                        {'name': 'round_winner', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
               frst ={
                    'attributes':[
                        {'name': 'ct_defuse_kits', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
               scnd ={
                    'attributes':[
                        {'name': 'ct_defuse_kits', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'}
               )


clm_one.print_summary()
clm_one.print_rulelist()
clm_one.print_rule(2)
clm_one.print_rule(3)
```
